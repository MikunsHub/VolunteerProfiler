# I see a problem of double indexing and random strings in the NAME,EMAIL & PHONE NUMBER columns indicating that those sensitive informations are being protected.

```python
#So we would be dropping the "Unnamed: 0" column
data = data.drop(["Unnamed: 0"],axis=1)
data.head()
```

# By glancing through this data, we can see how the "WHICH TEAM WOULD YOU LIKE TO JOIN" is randomly ordered, In my opinion the column name is too long and can be shortened to "Teams"

```python
#renaming the column of interest
data.rename(columns = {"WHICH TEAM WOULD YOU LIKE TO JOIN":"Teams"}, inplace = True)
```

```python
data.sort_values(by="Teams",axis=0,inplace=True)
data.head()
```

## As the idea of this process is to ensure each Team leader has a list of their respective volunteers to interview, it is necessary to extract the list.

```python
#the function below will do the necessary extraction
def sep_teams(TeamName):
    temp = data[((data.Teams == TeamName))]
    temp.to_csv(f"C:/Users/HP PC/Desktop/work/Notebooks/{TeamName}.csv")
    return temp
```
