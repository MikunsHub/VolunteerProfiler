```python
#importing libraries
#from __future__ import print_function
from mailmerge import MailMerge
import pandas as pd 
import numpy as np
```

```python
#reading the data
data = pd.read_csv(r'C:/Users/HP PC/Desktop/work/Notebooks/Publicity.csv')
```

```python
data.head()
```

```python
#read template file
template = "C:/Users/HP PC/Desktop/work/Notebooks/Template.docx"
```

```python
#loop through the dataset, check every column,pick the value in each column and initialize with the MailMerge template
document = MailMerge(template)
for i in range(len(data.index)):
    document.merge(Teams = str(data.iloc[i,7]), TELL = str(data.iloc[i,11]), DESCRIBE = str(data.iloc[i,12]), 
                        EMAIL = str(data.iloc[i,5]), HIGHLIGHT = str(data.iloc[i,10]), PHONE = str(data.iloc[i,6]), 
                        Response1 = str(data.iloc[i,8]), NAME = str(data.iloc[i,1]),GENDER = str(data.iloc[i,2]),DEPARTMENT = str(data.iloc[i,3]),PLEASE = str(data.iloc[i,14]),LEVEL = str(data.iloc[i,4]), Contribution = str(data.iloc[i,10]))
        
    file_name = str(data.iloc[i,7]) + 'tested' + '.docx'
        
    #create the document
    document.write(file_name)
```
